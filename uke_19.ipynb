```python
import numpy as np
import matplotlib.pyplot as plt

# Funksjon
def F(x,y):
    return np.sin(x) - y

# Start og slutt
x0 = 0
y0 = 1
x_slutt = 5

# Oppdeling
N = 500
h = (x_slutt - x0)/N

# Metoden
x_vektor = np.zeros((N))
y_vektor = np.zeros((N))
x = x0
y = y0
for n in range(0, N):
  x_vektor[n] = x
  y_vektor[n] = y
  y = y + F(x,y)*h
  x = x + h

# Plotter resultat
plt.plot(x_vektor, y_vektor)
```
