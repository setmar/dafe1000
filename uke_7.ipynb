```python
# Skript som plottar den deriverte

import numpy as np
import matplotlib.pyplot as plt

# Funksjonen
def f(x):
    """Funksjonen som skal deriverast"""
    return np.sqrt(x)

# Argument-verdi
a = 1

# Vektor med ulike steglengder
h = 0.75**np.arange(1, 10)    # Grov oppdeling
print(h)

# Tre estimat: framover-, bakover- og midtpunktsformel
fram_formel = (f(a+h)-f(a))/h
bak_formel = (f(a)-f(a-h))/h
midt_formel = (f(a+h)-f(a-h))/(2*h)

# Plottar feilen
plt.plot(h, fram_formel, 'b-')
plt.plot(h, bak_formel, 'g--')
plt.plot(h, midt_formel, 'm-.')
plt.show()    
```

```python
# Skript som plottar feilen i ulike estimat for
# den deriverte.

# Funksjonen - og den eksakte deriverte
def f(x):
    """Funksjonen som skal deriverast"""
    return np.sin(x**2)
    
def f_d(x):
    """Den deriverte"""
    return 2*x*np.cos(x**2)

# Argument-verdi
a = 1

# Eksakt svar
derivert = f_d(a)

# Vektor med ulike steglengder
h = 0.25**np.arange(1, 8)    # Grov oppdeling

# Tre estimat: framover-, bakover- og midtpunktsformel
fram_formel = np.abs((f(a+h)-f(a))/h - derivert)
bak_formel = np.abs((f(a)-f(a-h))/h - derivert)
midt_formel = np.abs((f(a+h)-f(a-h))/(2*h) - derivert)

# Plottar feilen
plt.plot(h, fram_formel, 'b-')
plt.plot(h, bak_formel, 'g--')
plt.plot(h, midt_formel, 'm-.')
plt.show()    
```

```python
"""Skript som plottar veksten i folketal i Oslo fr 2004 til 2018.
Den brukar midtpunktsformelen for numerisk derivasjon.

Kode tilhrende eksempel 4
"""

import numpy as np
import matplotlib.pyplot as plt

# Inputs
aar = np.arange(2004, 2019, 2)
folk = np.array([526887, 543514, 567980, 593045, 618683, 634293, 657478, 673469])

# Finn lengda
L = len(aar)

aar_midt = np.zeros(L-1)
fart_midt = np.zeros(L-1)

# Estimerar deriverte
for n in range(0, L-1):
    aar_midt[n] = (aar[n] + aar[n+1])/2        # Midtpunkt
    fart_midt[n] = (folk[n+1] - folk[n])/(2*1) # Estimerar farten

# Plottar vekstfarten
plt.plot(aar_midt, fart_midt, 'kx-', linewidth=2)
plt.grid(True)
plt.xlabel('r')
plt.ylabel('Fart (personer per r)')
```
