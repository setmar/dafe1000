Hei, dette er en tekstcelle med en liste:

- En ting
- En annen ting

```python
# Her skriver vi noe Python-kode

import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]

y = [10, 20, 30, 40, 50]

plt.plot(x, y)

def f(x):
    return x*10

print(f(x))

y2 = []
for i in range(0, len(x)):
    y2.append(f(x[i]))

print(y2)
plt.show()
plt.plot(x, y2)

x2 = np.linspace(1, 5, 5)
print(x2)

plt.show()
plt.plot(x2, f(x2))
```

```python
x = np.linspace(-1, 1, 20)

plt.plot(x, x)
plt.plot(x, np.exp(x))
plt.plot(x, np.log(x))

x = 123
print(np.log(np.exp(x)))

plt.gca().set_aspect('equal')
plt.grid()
```
