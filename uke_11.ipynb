```python
from sympy import var, plot_implicit
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)

def f(x):
    return 15**(1/4)-(2/15**(3/4))*(x-0.5)

fig, ax = plt.subplots()

plt.plot(x, f(x))
var('x y')

plot_implicit(x**4+(y/2)**4-1, fig=fig, ax=ax)
```
